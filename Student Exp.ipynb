```python
def freq_str(s,target):
    count=0
    for i in s:
        if target==i:
            count+=1
    return count
```

```python
freq_str('abababbabaaaaaa','b')
```

```python
def fun(variable):
    a=list()
    for i in variable:
        
        if i=='' or i==None:
            a.append('')
        else:
            a.append(freq_str(i,';')+1)
    return a
#     df[variable].replace(a,inplace=True)
#     variable = a.values
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def split(s):
    avg=s.split(' - ')
    avg = list(map(int,avg))
    avg=np.mean(avg)
#     print(avg)
    return avg
```

```python
split('200 - 500')
```

```python
def wow(variable):
    b=list()
    for i in variable:
        b.append(split(i))
    return b
```
